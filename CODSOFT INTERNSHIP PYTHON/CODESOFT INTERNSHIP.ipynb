```python
# QUESTION 1 

from IPython.display import clear_output

def displayList(toDoList):
    if(len(toDoList) == 0):
        print("YOU HAVE NO TASKS AT THE MOMENT!\n")
    else:
        print("YOUR CURRENT TO DO LIST: \n")
        for i in toDoList:
            print(f"{i[0]}: {i[1]}")
        print("\n")
            

toDoList = []
currentCount = 0

end = False

while(end != True):
    clear_output()
    displayList(toDoList)
    
    option = int(input("ENTER 1 IF YOU WANT TO ADD SOME TASK\n\nENTER 2 IF YOU WANT TO DELETE SOME TASK\n\nENTER 3 IF YOU WANT TO UPDATE SOME TASK\n\nENTER 4 TO END THE APPLICATION:\n\n"))

    clear_output()
    if(option == 1):
        task = input("ENTER THE TASK TO ADD IT TO YOUR TO DO LIST: \n\n\n")
        currentCount += 1
        toDoList.append([currentCount, task]) 
        
    elif(option == 2):
        
        if(len(toDoList) == 0):
            print("THERE ARE CURRENTLY NO ITEMS IN THE LIST! \n")
        
        else:
            displayList(toDoList)
            valid = False
            while(valid != True):
                taskNo = int(input("ENTER THE TASK NUMBER TO DELETE IT FROM YOUR TO DO LIST: \n"))
                if(taskNo > 0 and taskNo <= currentCount):
                    valid = True
                else:
                    print("INVALID INPUT!\n")
            removed = False
            for i in toDoList:
                if(taskNo == i[0]):
                    toDoList.remove(i)
                    currentCount -= 1
                    removed = True
                    
            for i in range(0, currentCount):
                toDoList[i][0] = i+1
        

            if(removed == False):
                print("TASK NUMBER DOES NOT EXIST!\n")
                
    elif(option == 3):
        
        if(len(toDoList) == 0):
            print("THERE ARE CURRENTLY NO ITEMS IN THE LIST!\n")
        
        else:
            displayList(toDoList)
            valid = False
            while(valid != True):
                taskNo = int(input("ENTER THE TASK NUMBER TO UPDATE IT IN YOUR TO DO LIST:\n"))
                if(taskNo > 0 and taskNo <= currentCount):
                    valid = True
                else:
                    print("INVALID INPUT!\n")
                    
            newTask = input("ENTER THE UPDATED TASK INFORMATION: \n")
            for i in toDoList:
                if(taskNo == i[0]):
                    toDoList[taskNo-1][1] = newTask
            print("TASK WAS UPDATED SUCCESSFULLY!\n")

    elif(option == 4):
        end = True
        print("APPLICATION ENDED SUCCESSFULLY!\n")
        
    else:
        print("INVALID INPUT, CHOOSE AGAIN!\n")
        
    
    

    
    
    
    
    
    
```

```python
#PASSWORD GENERATOR

def generatePassword(length):
    finalPw = ''
    for i in range(0, length):
        thisChar = random.randint(97,122)
        thisChar = chr(thisChar)
        finalPw += thisChar
    return finalPw
        

import random
correct = False
while(correct != True):
    length = int(input("INPUT THE LENGTH OF YOUR PASSWORD: \n\n"))
    
    if(length >= 0):
        correct = True
        
    else:
        print("\nINVALID INPUT, CHOOSE AGAIN!\n")
        input("PRESS ENTER TO CONTINUE...\n\n")
        clear_output()
        
        
        
        
    
print("\nFINAL GENERATED PASSWORD WITH RANDOM CHARACTERS: ", generatePassword(length))
    
    


```
