```python
# TO DO LIST

from IPython.display import clear_output #for better presenting of the output

#displays the to do list
def displayList(toDoList):
    clear_output()
    if(len(toDoList) == 0):
        print("YOU HAVE NO TASKS AT THE MOMENT!\n")
    else:
        print("YOUR CURRENT TO DO LIST: \n")
        for i in toDoList:
            print(f"{i[0]}: {i[1]}")
        print("\n")
            

toDoList = []
currentCount = 0

end = False

while(end != True):
    displayList(toDoList)
    option = int(input("ENTER 1 IF YOU WANT TO ADD SOME TASK\n\nENTER 2 IF YOU WANT TO DELETE/MARK AS DONE TASK\n\nENTER 3 IF YOU WANT TO UPDATE SOME TASK\n\nENTER 4 TO END THE APPLICATION:\n\n"))

    #adds item into the list
    if(option == 1):
        clear_output()
        task = input("ENTER THE TASK TO ADD IT TO YOUR TO DO LIST: \n\n")
        currentCount += 1
        toDoList.append([currentCount, task]) 

    #deletes or marks as done
    elif(option == 2):
        
        if(len(toDoList) == 0):
            print("\n THERE ARE CURRENTLY NO ITEMS IN THE LIST! \n")
            input("PRESS ENTER TO CONTINUE...\n\n")
            clear_output()
        
        else:
            valid = False
            while(valid != True):
                displayList(toDoList)
                taskNo = int(input("ENTER THE TASK NUMBER TO DELETE/MARK AS DONE FROM YOUR TO DO LIST: \n\n"))
                if(taskNo > 0 and taskNo <= currentCount):
                    valid = True
                else:
                    print("\nINVALID INPUT!\n")
                    input("PRESS ENTER TO CONTINUE...\n\n")
                    clear_output()
                    
            removed = False
            for i in toDoList:
                if(taskNo == i[0]):
                    toDoList.remove(i)
                    currentCount -= 1
                    removed = True
                    
            for i in range(0, currentCount):
                toDoList[i][0] = i+1

            print("\nTASK WAS DELETED SUCCESSFULLY!\n")
            input("PRESS ENTER TO CONTINUE...\n\n")
            clear_output()
        

            if(removed == False):
                print("TASK NUMBER DOES NOT EXIST!\n")
                input("PRESS ENTER TO CONTINUE...\n\n")
                clear_output()

    #updates contents of some task
    elif(option == 3):
        
        if(len(toDoList) == 0):
            print("\nTHERE ARE CURRENTLY NO ITEMS IN THE LIST!\n")
            input("PRESS ENTER TO CONTINUE...\n\n")
            clear_output()
        
        else:
            
            valid = False
            while(valid != True):
                displayList(toDoList)
                taskNo = int(input("ENTER THE TASK NUMBER TO UPDATE IT IN YOUR TO DO LIST:\n\n"))
                if(taskNo > 0 and taskNo <= currentCount):
                    valid = True
                else:
                    print("\nINVALID INPUT!\n")
                    input("PRESS ENTER TO CONTINUE...\n\n")
                    clear_output()

            clear_output()
            newTask = input("\nENTER THE UPDATED TASK INFORMATION: \n\n")
            for i in toDoList:
                if(taskNo == i[0]):
                    toDoList[taskNo-1][1] = newTask
            print("\nTASK WAS UPDATED SUCCESSFULLY!\n")
            input("PRESS ENTER TO CONTINUE...\n\n")
            clear_output()

    #ends the program
    elif(option == 4):
        end = True
        print("\nAPPLICATION ENDED SUCCESSFULLY!\n")
        
        
    else:
        print("\nINVALID INPUT, CHOOSE AGAIN!\n")
        input("PRESS ENTER TO CONTINUE...\n\n")
        clear_output()
        
    
    

    
    
    
    
    
    
```

```python
# ROCK PAPER SCISSORS GAME

#random turn by the computer
def computerGuess(choices):
    return choices[random.randint(0,2)] 

from IPython.display import clear_output
import random

choices = ['rock', 'paper', 'scissors']
userScore = 0
computerScore = 0
end = False

while(end != True):
    correct = False
    while(correct != True):
        clear_output()
        userChoice = int(input("ENTER 1 TO CHOOSE ROCK\n\nENTER 2 TO CHOOSE PAPER\n\nENTER 3 TO CHOOSE SCISSORS: \n\n"))

        if(userChoice == 1):
            userChoice = 'rock'
            correct = True
        elif(userChoice == 2):
            userChoice = 'paper'
            correct = True
        elif(userChoice == 3):
            userChoice = 'scissors'
            correct = True
        else:
            print("\nINVALID INPUT! \n")
            input("PRESS ENTER TO CONTINUE...\n\n")
            clear_output()

    clear_output()
    print("USER SELECTED: ", userChoice, "\n")
    computerChoice = computerGuess(choices)
    print("COMPUTER SELECTED: ", computerChoice, "\n")
    
    if(userChoice == computerChoice):
        print("GAME IS A TIE!\n")
        input("PRESS ENTER TO CONTINUE...\n\n")
        clear_output()
        
        userScore += 100
        computerScore += 100
    
    elif(userChoice == 'rock' and computerChoice == 'paper'):
        print("COMPUTER WINS!\n")
        end = True
        computerScore += 100
    
    elif(userChoice == 'rock' and computerChoice == 'scissors'):
        print("YOU WIN!\n")
        end = True
        userScore += 100
                           
    elif(userChoice == 'paper' and computerChoice == 'rock'):
        print("YOU WIN!\n")
        end = True
        userScore += 100
                           
    elif(userChoice == 'scissors' and computerChoice == 'rock'):
        print("COMPUTER WINS!\n")
        end = True
        computerScore += 100
    
    elif(userChoice == 'paper' and computerChoice == 'scissors'):
        print("COMPUTER WINS!\n")
        end = True
        computerScore += 100
                           
    elif(userChoice == 'scissors' and computerChoice == 'paper'):
        print("YOU WIN!\n")
        end = True
        userScore += 100
        
    print("COMPUTER SCORE: ", computerScore, "\n")
    print("USER SCORE: ", userScore, "\n")

    
    if(end == True):
        correct = False
        while(correct != True):
                option = int(input("DO YOU WANT TO PLAY AGAIN? ENTER 1 FOR YES AND ENTER 2 FOR NO: \n\n"))
                if(option == 1):
                    end = False
                    correct = True
                elif(option == 2):
                    print("\nGAME TERMINATED SUCCESSFULLY!")
                    break
                else:
                    print("\nINVALID INPUT!\n")
                    input("PRESS ENTER TO CONTINUE...\n\n")
                    clear_output()
                    
        
        
    

```

```python
#PASSWORD GENERATOR

#takes length as input and generates pw of that length
def generatePassword(length):
    finalPw = ''
    for i in range(0, length):
        thisChar = random.randint(97,122)
        thisChar = chr(thisChar)
        finalPw += thisChar
    return finalPw
        

import random
correct = False
while(correct != True):
    length = int(input("INPUT THE LENGTH OF YOUR PASSWORD: \n\n"))
    
    if(length >= 0):
        correct = True
        
    else:
        print("\nINVALID INPUT, CHOOSE AGAIN!\n")
        input("PRESS ENTER TO CONTINUE...\n\n")
        clear_output()
        
        
        
        
    
print("\nFINAL GENERATED PASSWORD WITH RANDOM CHARACTERS: ", generatePassword(length))
    
    


```
